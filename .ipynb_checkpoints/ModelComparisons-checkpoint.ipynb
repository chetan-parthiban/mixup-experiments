```python
import torch
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim
from torch.utils.data import DataLoader

import torchvision
from torchvision.datasets import ImageFolder
from torchvision import transforms

from tqdm import tqdm, trange
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
from Utilities import *
from TrainingLoops import *
```

```python
trainer = ModelManager(train_dir = 'data/imagenette2-320/train', val_dir = 'data/imagenette2-320/val', n_classes = 10, epochs = 5, img_size = 224)
```

```python
linear_results = runNTimes(10, trainer.train_linear)
```

```python
unaugmented_results = runNTimes(10, trainer.train_deep_unaugmented)
```

```python
augmented_results = runNTimes(10, trainer.train_deep_augmented)
```

```python
mixup_results = runNTimes(10, trainer.train_mixup)
```

```python
linear_performance = [result['final_perf'] for result in linear_results]
print('Linear Performance | mean =', np.mean(linear_performance), '| std =', np.std(linear_performance))
unaugmented_performance = [result['final_perf'] for result in unaugmented_results]
print('Unaugmented Performance | mean =', np.mean(unaugmented_performance), '| std =', np.std(unaugmented_performance))
augmented_performance = [result['final_perf'] for result in augmented_results]
print('Augemented Performance | mean =', np.mean(augmented_performance), '| std =', np.std(augmented_performance))
mixup_performance = [result['final_perf'] for result in mixup_results]
print('mixup Performance | mean =', np.mean(mixup_performance), '| std =', np.std(mixup_performance))
```

```python
from scipy.stats import ttest_ind
print('mixup performance vs linear performance p =',ttest_ind(mixup_performance, linear_performance).pvalue)
print('mixup performance vs unaugmented performance p =',ttest_ind(mixup_performance, unaugmented_performance).pvalue)
print('mixup performance vs augmented performance p =',ttest_ind(mixup_performance, augmented_performance).pvalue)
```

```python
plt.title('Training Losses')
plt.ylabel('Loss')
plt.xlabel('Number of Iterations')
plt.ylim(0,3)
plt.plot(linear_results[3]['losses'])
plt.plot(unaugmented_results[3]['losses'])
plt.plot(augmented_results[3]['losses'])
plt.plot(mixup_results[3]['losses'])
plt.legend(['Linear', 'Deep Learning', 'Augmented Deep Learning', 'Mixup'])

plt.show()

plt.title('Validation Accuracies')
plt.ylim(0,100)
plt.ylabel('Accuracy')
plt.xlabel('Epochs')
plt.plot([x* 100 for x in linear_results[3]['val_accs']])
plt.plot([x* 100 for x in unaugmented_results[3]['val_accs']])
plt.plot([x* 100 for x in augmented_results[3]['val_accs']])
plt.plot([x* 100 for x in mixup_results[3]['val_accs']])
plt.legend(['Linear', 'Deep Learning', 'Augmented Deep Learning', 'Mixup'])

plt.show()
```

```python
perf_means = [np.mean(linear_performance), np.mean(unaugmented_performance), np.mean(augmented_performance), np.mean(mixup_performance)]
perf_errs = [np.std(linear_performance), np.std(unaugmented_performance), np.std(augmented_performance), np.std(mixup_performance)]

plt.title('Final Performance Comparison w/ Error Bars')
plt.bar([0],perf_means[0], yerr=perf_errs[0])
plt.bar([1],perf_means[1], yerr=perf_errs[1])
plt.bar([2],perf_means[2], yerr=perf_errs[2])
plt.bar([3],perf_means[3], yerr=perf_errs[3])
plt.legend(['Linear', 'Deep Learning', 'Augmented Deep Learning', 'Mixup'],loc = 'lower right')
plt.show()
```
